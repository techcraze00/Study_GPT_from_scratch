### Step 1

Write the `BasicTokenizer` class, with the following three core functions:

- `def train(self, text, vocab_size, verbose=False)`
- `def encode(self, text)`
- `def decode(self, ids)`


````python
from typing import List, Dict
import collections

class BasicTokenizer:
    def __init__(self):
        self.training_data = r"test data\taylorswift.txt"
        self.merged_ids = {} # Initialize consistently
        self.vocab = {}
        
    def train(self, data, vocab_size, verbose=False):
        # 1. Inner Helper Functions (no 'self' as param)
        def get_pairs(tokens):
            counts = collections.defaultdict(int)
            for pair in zip(tokens, tokens[1:]):
                counts[pair] += 1
            return counts
            
        def merge_ids(ids, pair, idx):
            new_ids = []
            i = 0
            while i < len(ids):
                # Check for the specific pair (p0, p1)
                if i < len(ids) - 1 and ids[i] == pair[0] and ids[i+1] == pair[1]:
                    new_ids.append(idx)
                    i += 2
                else:
                    new_ids.append(ids[i])
                    i += 1
            return new_ids

        # 2. Prepare Data
        file_path = data if data else self.training_data
        with open(file_path, "r", encoding="utf-8") as f:
            text = f.read()
        
        # Convert to initial byte IDs
        ids = list(text.encode("utf-8"))
        
        # 3. Training Loop
        num_merges = vocab_size - 256
        for i in range(num_merges):
            stats = get_pairs(ids)
            if not stats: break
            
            top_pair = max(stats, key=stats.get)
            new_idx = 256 + i
            
            if verbose: print(f"Merging {top_pair} -> {new_idx}")
            
            ids = merge_ids(ids, top_pair, new_idx)
            self.merged_ids[top_pair] = new_idx

        # 4. Build Vocabulary
        self.vocab = {i: bytes([i]) for i in range(256)}
        for (p0, p1), idx in self.merged_ids.items():
            self.vocab[idx] = self.vocab[p0] + self.vocab[p1]
        
        return self.vocab   

    def encoder(self, text: str):
        # Use instance's merged_ids directly
        ids = list(text.encode("utf-8"))
        while len(ids) >= 2:
            # Re-calculating stats for the current sequence
            stats = {}
            for pair in zip(ids, ids[1:]):
                stats[pair] = stats.get(pair, 0) + 1
            
            # Find which of the current pairs was merged earliest in training
            # We want the pair with the smallest 'new_idx' in self.merged_ids
            pair = min(stats.keys(), key=lambda p: self.merged_ids.get(p, float("inf")))
            
            if pair not in self.merged_ids:
                break # No more mergeable pairs
                
            idx = self.merged_ids[pair]
            # Reuse the merge logic (could be moved to a static method)
            ids = self._merge_ids_util(ids, pair, idx)
        return ids

    def _merge_ids_util(self, ids, pair, idx):
        # Helper for internal use in both train and encoder
        new_ids = []
        i = 0
        while i < len(ids):
            if i < len(ids) - 1 and ids[i] == pair[0] and ids[i+1] == pair[1]:
                new_ids.append(idx)
                i += 2
            else:
                new_ids.append(ids[i])
                i += 1
        return new_ids
    
    def decoder(self, ids):
        tokens = b"".join(self.vocab[i] for i in ids)
        return tokens.decode("utf-8", errors="replace")
````

````python
# Runner function"

tokenizer = BasicTokenizer()
tokenizer.train(data=None, vocab_size=32)

# print(tokenizer.vocab)

encoding = tokenizer.encoder("hello ??")
print(tokenizer.decoder(encoding) == "hello ??")
````
